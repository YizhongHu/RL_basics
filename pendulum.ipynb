```python
import gym

from DDPG import DeepDeterministicPolicyGradient

from pendulum_wrappers import RewardEnergy
```

```python
model = DeepDeterministicPolicyGradient(actor_dims=[3, 32, 32, 1], critic_dims=[4, 32, 32, 1], action_bounds=[[-2, 2]],
                                        exploration=1, exploration_decay=.01, gamma=.95, 
                                        memory=100000, start_updating=1000, batch_size=32,
                                        actor_lr=1e-4, critic_lr=1e-4)
with gym.make('Pendulum-v1') as env:
    env = RewardEnergy(env)
    model.train(env, episodes=1000)
```

```python
with gym.make('Pendulum-v1') as env:
    iters, total_reward = model.simulate(env, render=True, evaluation=True)
print(f'Iters: {iters}, Total Reward: {total_reward}')
```
